```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\SUVARCHALA\\Downloads\\Dataset .csv")
```

```python
# Display the column names
print(df.columns)
```

**# Percentage of restaurants offering table booking**

```python
# Percentage of restaurants offering table booking
table_booking_percent = df['Has Table booking'].value_counts(normalize=True) * 100
print('\nPercentage of restaurants offering table booking:\n', table_booking_percent)
```

```python
# Percentage of restaurants offering online delivery
online_delivery_percent = df['Has Online delivery'].value_counts(normalize=True) * 100
print('\nPercentage of restaurants offering online delivery:\n', online_delivery_percent)
```

**# Average ratings for table booking**

```python

# 2. Compare the average ratings of restaurants with table booking and those without

# Average ratings for table booking
avg_rating_table_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()
print('\nAverage ratings of restaurants with and without table booking:\n', avg_rating_table_booking)

plt.figure(figsize=(8, 6))
sns.barplot(x=avg_rating_table_booking.index, y=avg_rating_table_booking.values, palette='viridis')
plt.title('Average Ratings of Restaurants With and Without Table Booking')
plt.xlabel('Table Booking')
plt.ylabel('Average Aggregate Rating')
plt.show()
```

**# Availability of online delivery among different price ranges**

```python
# 3. Analyze the availability of online delivery among restaurants with different price ranges

# Availability of online delivery among different price ranges
online_delivery_price_range = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
print('\nAvailability of online delivery among different price ranges:\n', online_delivery_price_range)

plt.figure(figsize=(10, 6))
online_delivery_price_range.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Availability of Online Delivery Among Different Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Percentage')
plt.show()
```
